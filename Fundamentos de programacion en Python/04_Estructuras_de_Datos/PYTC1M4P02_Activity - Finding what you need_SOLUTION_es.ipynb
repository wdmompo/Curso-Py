El código en la celda de abajo hace exactamente eso! Antes de ejecutarlo, desglosémoslo:

* **Funciones Auxiliares:**  En programación, una función auxiliar es simplemente una función que ayuda a otra función a completar su tarea.  Piensa en ellas como pequeños asistentes de bibliotecario.  Realizan tareas específicas para facilitar el trabajo principal (ordenar el catálogo).
    * `get_title(book)`:  Esta auxiliar simplemente extrae el título del diccionario de un libro, para que podamos centrarnos solo en eso al ordenar.
    * `sort_catalog_by_title(catalog)`: Este es el bibliotecario principal encargado de ordenar.  Utiliza un truco especial de Python (`catalog.sort()`) para organizar todos los libros en orden alfabético por sus títulos.

* **Poniéndolo Todo Junto:**
    * Llamamos a la función `sort_catalog_by_title` para hacer la ordenación real.

```python
# Helper function 1: 
def get_title(book): 
  """Helper function to extract the title from a book dictionary."""
  return book['title']

# Helper function 2: 
def sort_catalog_by_title(catalog): 
  """Sorts the book catalog alphabetically by title."""
  catalog.sort(key=get_title)
    
# Sort the catalog
sort_catalog_by_title(book_catalog)

# Display the sorted catalog
for book in book_catalog:
  print(f"Title: {book['title']}, Author: {book['author']}, Publication Year: {book['publication_year']}")
```

Pero, ¿qué pasa si quieres encontrar un libro sin escanear toda la lista? Ahí es donde los algoritmos de búsqueda son útiles. Ellos te ayudan a localizar elementos específicos dentro de un conjunto de datos, justo como un bibliotecario te ayuda a encontrar un libro en la estantería.


En la siguiente celda, explorarás uno de estos algoritmos llamado **búsqueda lineal**. Es como revisar sistemáticamente cada libro en una estantería hasta encontrar el que estamos buscando. Pero, ¿cómo se traduce revisar cada libro en una estantería uno por uno en código Python?

Vamos a desglosar el código en la siguiente celda antes de ejecutarlo:

La función `search_books` a continuación hace lo siguiente:

1. **Recibe entrada:** Necesita dos cosas para funcionar:
    * El `catálogo`:  Esta es nuestra lista ordenada de diccionarios de libros.
    * La `consulta`:  Esto es lo que estamos buscando – puede ser un título de libro o un nombre de autor. 


2. **Configura una lista para resultados:** Crea una lista vacía llamada `results` para almacenar cualquier libro que coincida con nuestra búsqueda. 

3. **Revisa cada libro:**  Recorre cada `libro` en el `catálogo`:
    * Convierte tanto la `consulta` como el `título` y el `autor` del libro a minúsculas. Esto hace que la búsqueda no distinga entre mayúsculas y minúsculas (por lo tanto, "el gran gatsby" y "El Gran Gatsby" se consideran iguales). 
    * Verifica si la `consulta` se encuentra dentro del `título` del libro O del `autor` del libro.
    * Si hay una coincidencia, añade ese `libro` a la lista `results`. 


4. **Devuelve los resultados:**  Finalmente, nos devuelve la lista `results`, que contiene todos los libros que coincidieron con nuestra búsqueda (o una lista vacía si no se encontró ninguno).

**En términos más simples, es como recorrer cada libro en la estantería, verificar si es el que queremos, y ponerlo a un lado si lo es.**

**¿Listo para verlo en acción? ¡Ejecuta la celda de abajo!**


```python
# Helper function: Linear Search
def search_books(catalog, query):
  """Searches for books by title or author using linear search."""
  results = []
  for book in catalog:
    if query.lower() in book['title'].lower() or query.lower() in book['author'].lower():
      results.append(book)
  return results
```

Por suerte, hay una forma más rápida cuando nuestro catálogo está ordenado: **búsqueda binaria**. Es como jugar a un juego de adivinanzas donde sigues dividiendo las posibilidades a la mitad hasta encontrar lo que buscas. Pero, ¿cómo se traduce esto en código Python?

Vamos a desglosar el código en la siguiente celda antes de ejecutarlo:

Aquí está lo que hace la función `binary_search_books`:

1. **Toma entrada:**  Al igual que la búsqueda lineal, necesita el `catálogo` y la `consulta` (el título del libro que estamos buscando).

2. **Establece límites:**  Comienza marcando el inicio (`low`) y el final (`high`) del catálogo.

3. **El juego de adivinanzas:**
   * Calcula el punto medio (`mid`) entre `low` y `high`.
   * Verifica si el libro en el punto `mid` es el que buscamos.
   * Si no, compara el título del libro en el punto `mid` con nuestra `consulta`:
     * Si el título en el `mid` es "menor que" nuestra `consulta`, sabemos que el libro que queremos debe estar en la mitad *superior* del catálogo, así que ajustamos `low` a `mid + 1`.
     * Si el título en el `mid` es "mayor que" nuestra `consulta`, sabemos que el libro que queremos debe estar en la mitad *inferior*, así que ajustamos `high` a `mid - 1`.
   * Continúa repitiendo este proceso, reduciendo el área de búsqueda a la mitad cada vez, hasta encontrar el libro o determinar que no está allí.

4. **Devuelve el resultado:**  Si encuentra el libro, devuelve todo el diccionario del libro. De lo contrario, devuelve `None` para indicar que no se encontró el libro.

**Piensa en esto así:** Estás buscando una palabra en un diccionario. En lugar de comenzar desde el principio, lo abres en algún lugar del medio. Si la palabra que quieres aparece *después* de las palabras en esa página, pasas a la segunda mitad del diccionario. Si aparece *antes*, pasas a la primera mitad. ¡Sigues haciendo esto hasta encontrar la página exacta!

**¿Listo para verlo en acción? ¡Ejecuta la celda de abajo!**


```python
def binary_search_books(catalog, query):
  """Searches for books by title using binary search (assuming sorted catalog)."""
  low = 0
  high = len(catalog) - 1

  while low <= high:
    mid = (low + high) // 2
    if catalog[mid]['title'].lower() == query.lower():
      return catalog[mid]
    elif catalog[mid]['title'].lower() < query.lower():
      low = mid + 1
    else:
      high = mid - 1

  return None  # Book not found
```

En las celdas anteriores, usaste búsqueda lineal para encontrar "El Gran Gatsby", pero ¿puedes encontrarlo incluso *más rápido*?  

Recuerda, la búsqueda binaria es súper rápida cuando nuestros datos están ordenados. Funciona dividiendo repetidamente el área de búsqueda a la mitad, como un juego de adivinanzas de alta tecnología.

```python
# This is like using a stopwatch to see how long it takes to find the books. 
# We start the timer, do the search, then stop the timer and calculate how much time passed.
import time

# Loads the big_book_catalog which has which has ~271,380 rows!
big_book_catalog_df = pd.read_csv('big_book_catalog.csv', low_memory=False)

# Convert 'title' and 'author' to strings, handle NaN
big_book_catalog_df['title'] = big_book_catalog_df['title'].fillna('').astype(str)
big_book_catalog_df['author'] = big_book_catalog_df['author'].fillna('').astype(str)

sorted_df = big_book_catalog_df.sort_values(by=['title'])
big_book_catalog = sorted_df.to_dict(orient='records')

# Search for "The Great Gatsby" using Binary Search
query = "The Great Gatsby"  # Example query

start_time = time.time()

search_results = binary_search_books(big_book_catalog, query)

elapsed_time_binary = time.time() - start_time

print(f"\nBinary search took {elapsed_time_binary:.5f} seconds.")

if search_results:
  print("\nBinary Search Result for 'The Great Gatsby':")
  print(f"Title: {search_results['title']}, Author: {search_results['author']}, Publication Year: {search_results['publication_year']}")
else:
  print("Binary Search: Book not found")
```
