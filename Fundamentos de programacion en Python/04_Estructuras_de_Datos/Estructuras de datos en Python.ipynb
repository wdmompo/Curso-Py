## Datos Estructurados



Los datos estructurados son información digital organizada en un formato estructurado. Permite a los ordenadores acceder, gestionar y procesar los datos más fácilmente.

Los datos estructurados suelen estructurarse mediante tablas u otras formas similares de categorizar la información.

<h3>Características, ventajas y desventajas de los datos estructurados</h3>

Los datos estructurados tienen algunas ventajas significativas sobre los no estructurados.

- Es más fácil buscar, almacenar y recuperar datos estructurados porque están estructurados de forma predecible y coherente.
Además, los ordenadores pueden manipular y analizar fácilmente los datos estructurados.
- Los datos estructurados también ocupan menos espacio de almacenamiento que los no estructurados debido a su formato estructurado.
- Sin embargo, los datos estructurados también pueden ser rígidos e inflexibles. Es difícil añadir nuevos elementos o cambiar los existentes en los datos estructurados, ya que tienen una estructura fija.

<h3>¿Cuáles son los tipos de datos estructurados?</h3>

Los tipos de datos estructurados más utilizados son el texto, los números, las fechas y los valores booleanos.

- Los datos estructurados de texto incluyen palabras o frases que pueden buscarse en una base de datos.
- Los datos estructurados numéricos incluyen valores numéricos como precios o códigos de producto.
- Los datos estructurados de fechas se utilizan para almacenar información sobre la fecha y la hora en que ocurrió algo.
- Los datos estructurados booleanos se refieren a valores verdaderos o falsos, como si un artículo está en stock o no.


<h4> Ejemplos de datos estructurados </h4>

Los datos estructurados se utilizan en numerosas aplicaciones. Algunos ejemplos comunes son:

- Bases de datos, como las relacionales bases de datos, que almacenan datos estructurados en tablas.
- Hojas de cálculo, como Microsoft Excel o Google Sheets, que son colecciones estructuradas de números y texto.
- Archivos de registro de servidores y aplicaciones que contienen información estructurada sobre usuarios y eventos.
- Sensores que generan flujos de datos estructurados para medir fenómenos físicos como la temperatura o la presión atmosférica.

## Datos No Estructurados



Los datos no estructurados son información digital que no tiene un formato estructurado. Suelen consistir en archivos de texto, imágenes, audio y vídeo.

<h3>Características, ventajas y desventajas de los datos no estructurados</h3>

Los datos no estructurados tienen algunas ventajas significativas sobre los estructurados.

- Son más flexibles y dinámicos que los datos estructurados porque no tienen una estructura fija. Esto facilita la adición de nuevos elementos o la modificación de los existentes en los datos no estructurados.
- Además, los datos no estructurados pueden contener una mayor variedad de información que los estructurados.
- Sin embargo, los datos no estructurados son más difíciles de almacenar y buscar que los estructurados porque no tienen un formato estructurado.

<h4>¿Cuáles son los tipos de datos no estructurados?</h4>

Los tipos más comunes de datos no estructurados son texto, audio, vídeo, imágenes y formatos de documentos estructurados como los PDF.

Ejemplos de datos no estructurados
Los datos no estructurados se utilizan en numerosas aplicaciones. Algunos ejemplos comunes son:

- Publicaciones en redes sociales, que contienen mensajes estructurados, imágenes, vídeos y otros tipos de contenido.
- Imágenes y vídeos, que suelen ser archivos no estructurados.
- Documentos de texto, como correos electrónicos o notas, que contienen texto estructurado.
- Grabaciones de voz que contienen habla estructurada.

## 4. Mas con bucles, listas, y tuplas

````python
equipo = [('Marta', 20, 'centro'),
          ('Ana', 22, 'base'),
          ('Gabi', 22, 'escolta'),
          ('Luz', 21, 'ala-pívot'),
          ('Lorena', 19, 'alero'),
            ]
````

````python
# Crea una función para extraer los nombres y las posiciones de la lista de jugadores y
# formatearlos para imprimirlos. Devuelve una lista.
def player_position(jugadores):
    resultado = []
    for nombre, edad, posicion in jugadores:
        resultado.append('Nombre:   {:<15} \nPosición: {:<15}\n'.format(nombre, posicion))
    return resultado
````

````python
# Recorre la lista de nombres y posiciones formateados producidos por la función
# player_position() e imprímelos.
for jugador in player_position(equipo):
    print(jugador)
````

````python
# Los bucles anidados pueden generar las diferentes combinaciones de puntos en un conjunto de dichas de domino.
for izquierda in range(7):
    for derecha in range(izquierda, 7):
        print(f"[{izquierda}|{derecha}]", end=" ")
    print('\n')
````

````python
# Crea una lista de dominós, donde cada ficha  está representado como una tupla.
fichas = []
for izquierda in range(7):
    for derecha in range(izquierda, 7):
        fichas.append((izquierda, derecha))

fichas
````

````python
# Select index 1 of the tuple at index 4 in the list of dominoes.
fichas[4][1]
````

````python
# Puedes usar un bucle for para sumar los puntos en cada dominó y agregar
# la suma a una nueva lista.
puntos_desde_bucle = []
for ficha in fichas:
    puntos_desde_bucle.append(ficha[0] + ficha[1])
print(puntos_desde_bucle)
````

````python
# Una comprensión de lista produce el mismo resultado con menos código.
# una lista por comprensión es una manera de hecer el bucle in situ en la lista
puntos_desde_comp_lista = [ ficha[0] + ficha[1] for ficha in fichas ]
puntos_desde_bucle == puntos_desde_comp_lista
````

## 5. Introducción a diccionarios

Un diccionario es una estructura de datos que permite almacenar y organizar información en pares clave-valor. A diferencia de las listas y las tuplas, que están indexadas numéricamente, los diccionarios utilizan claves únicas para acceder a sus valores asociados. Se definen utilizando llaves {} y contienen elementos en forma de clave: valor, separados por comas.

````python
# Crea un diccionario con los corrales como claves y los animales que contienen como valores.
# Los diccionarios se pueden instanciar usando llaves.

zoo = {
    'corral_1' : 'pingüinos',
    'corral_2' : 'cebras',
    'corral_3' : 'leones',
}
 # Seleccionar la clave  `corral_2`  devuelve  `cebras` , el valor almacenado en esa clave.

zoo['corral_2']
````

````python
print(zoo)
````

````python
# No puedes acceder a los valores de un diccionario por nombre utilizando corchetes
# porque la computadora lo interpreta como una clave, no como un valor.
zoo['cebras']
````

````python
# Los diccionarios también se pueden instanciar utilizando la función dict().

zoo = dict(
    pen_1 = 'monos',
    pen_2 = 'cebras',
    pen_3 = 'leones',
    )

zoo['pen_2']
````

````python
zoo
````

````python
# Otra forma de crear un diccionario utilizando la función dict().
zoo2 = dict(
    [
     ['pen_1', 'monos'],
     ['pen_2', 'cebras'],
     ['pen_3', 'leones'],
     [1 ,"uno"],
     [(0,0) , "par"],
     ["comprar", ["mandarina","peras"]]
    ]
)

zoo2
````

````python
zoo2["comprar"]
````

````python
zoo2["comprar"][1]
````

````python
# Asignar un nuevo par clave: valor a un diccionario existente.

zoo2['pen_4'] = 'cocodrilos'
zoo2
````

````python
zoo2["comprar"].append("Manzana")
````

````python
zoo2
````

````python
# Los diccionarios no están ordenados y no admiten indexación numérica.

zoo[2]
````

````python
# Use el operador `in` para evaluar si la clave existe o no en el diccionario.
print('pen_1' in zoo)
print('pen_7' in zoo)
````

## 7. Introducción a conjuntos

Un conjunto (set en inglés) es una colección no ordenada y sin elementos duplicados. Los conjuntos son útiles cuando necesitas almacenar una colección de elementos únicos y no te importa el orden en que se almacenan. Se definen utilizando llaves {} o la función set().

````python
# La función set() convierte una lista en un conjunto.

x = set(['foo', 'bar', 'baz', 'foo'])
print(x)
````

````python
dir(x)
````

````python
# La función set() convierte una tupla en un conjunto.
x = set(('foo','bar','baz', 'foo'))
print(x)
````

````python
# The set() function converts a string to a set.
x = set('foo')
print(x)
````

````python
# Puedes usar llaves para instanciar un conjunto.
x = {'foo'}
print(type(x))
 # Pero las llaves vacías están reservadas para los diccionarios.
y = {}
print(type(y))
````

````python
# Al instanciar un conjunto con llaves, los contenidos se tratan como literales.
x = {'foo', 'bar', 'baz', 'foo'}
print(x)
````

````python
# El método intersection() (&) devuelve los elementos comunes entre dos conjuntos.
conjunto1 = {1, 2, 3, 4, 5, 6}
conjunto2 = {4, 5, 6, 7, 8, 9}
print(conjunto1.intersection(conjunto2))
print(conjunto1 & conjunto2)
````

````python
# El método union() (|) devuelve todos los elementos de dos conjuntos, cada uno representado una vez.
x1 = {'foo', 'bar', 'baz'}
x2 = {'baz', 'qux', 'quux'}
print(x1.union(x2))
print(x1 | x2)
````

````python
# El método difference() (-) devuelve los elementos en set1 que no están en set2.
conjunto1 = {1, 2, 3, 4, 5, 6}
conjunto2 = {4, 5, 6, 7, 8, 9}
print(conjunto1.difference(conjunto2))
print(conjunto1 - conjunto2)
````

````python
# ... y los elementos en set2 que no están en set1.
print(conjunto2.difference(conjunto1 ))
print(conjunto2 - conjunto1 )
````

````python
# El método symmetric_difference() (^) devuelve todos los valores de cada conjunto que no están en ambos conjuntos.
conjunto1 = {1, 2, 3, 4, 5, 6}
conjunto2 = {4, 5, 6, 7, 8, 9}
conjunto2.symmetric_difference(conjunto1)
conjunto2 ^ conjunto1
````

````python
def proceso_txt(tx):
  diccionario = {"entidad_id"    : [],
               "nombre_entidad"  : [],
               "Doc_Debitos"     : [],
               "Importe_Debitos" : [],
               "Doc_Creditos"    : [],
               "Importe_Creditos": []}
  # con with me aseguro que el archivo se cierre cuando termine el bloque de codigo
  with open(tx,"r") as archivo:
    for lin , linea in enumerate(archivo):
      #tengo 12 lineas de encabezado que no me interezan hasta llegar a los datos
      if lin > 12 and not linea.startswith("-") and not linea.startswith("TOTALES"):
        datos = linea.split()
        #print(datos)
       df if len(datos) == 6:  # me aseguro de que tenga los 6 datos asi excluyo la liea final
          # agrego cada dato en su clave
          # puedo ya convertir los string en valores numericos
          diccionario['entidad_id'].append(datos[0])
          diccionario['nombre_entidad'].append(datos[1])
          diccionario['Doc_Debitos'].append(int(datos[2]))
          diccionario['Importe_Debitos'].append(float(datos[3]))
          diccionario['Doc_Creditos'].append(int(datos[4]))
          diccionario['Importe_Creditos'].append(float(datos[5]))

  df = pd.DataFrame(diccionario)
  return df
````

````python
def proceso_txt_2(tx):
  diccionario = {"entidad_id"    : [],
               "nombre_entidad"  : [],
               "Doc_Debitos"     : [],
               "Importe_Debitos" : [],
               "Doc_Creditos"    : [],
               "Importe_Creditos": [],
                "moneda" : []}
  # con with me aseguro que el archivo se cierre cuando termine el bloque de codigo
  with open(tx,"r") as archivo:
    for lin , linea in enumerate(archivo):
      if lin == 6:
        moneda = linea.split(":")[1].strip()
      #tengo 12 lineas de encabezado que no me interezan hasta llegar a los datos
      if lin > 12 and not linea.startswith("-") and not linea.startswith("TOTALES"):
        datos = linea.split()
        #print(datos)
        # if len(datos) == 6:  # me aseguro de que tenga los 6 datos asi excluyo la liea final
          # agrego cada dato en su clave
          # puedo ya convertir los string en valores numericos
        diccionario['entidad_id'].append(datos[0])
        diccionario['nombre_entidad'].append(" ".join(datos[1:-4]))
        diccionario['Doc_Debitos'].append(int(datos[-4]))
        diccionario['Importe_Debitos'].append(float(datos[-3]))
        diccionario['Doc_Creditos'].append(int(datos[-2]))
        diccionario['Importe_Creditos'].append(float(datos[-1]))
        diccionario["moneda"].append(moneda)

  df = pd.DataFrame(diccionario)
  return df
````

````python
def genera_archivos(num):
  for doc in range(num):
    with open("doc_"+str(doc)+".txt", "w") as archivo:
      archivo.write(encabezado.format("PROD01","PESOS"))
      #print(encabezado.format("PROD01","PESOS"))
      filas = random.randint(1,30)
      #archivo.write("-"*74)
      #print("-"*74)
      total_doc_d, total_deb, total_doc_c, total_cre = 0,0,0,0
      for fila in range(filas):

        importe_debitos  = round(random.random() * random.randint(1000,1000000),2)
        importe_creditos = round(random.random() * random.randint(1000,1000000),2)

        doc_debitos  = random.randint(0,50)
        doc_creditos = random.randint(0,50)

        total_deb += importe_debitos
        total_cre   += importe_creditos
        total_doc_d += doc_debitos
        total_doc_c += doc_creditos

        id_banco, banco = random.choice(bancos)

        fila = f"{id_banco:<6} {banco:<17}{doc_debitos:>2}{importe_debitos:>16.2f}{doc_creditos:>11}{importe_creditos:>17.2f}\n"

        archivo.write(fila)
        archivo.write("-"*74+"\n")

      archivo.write(f"TOTALES{total_doc_d:>19}{total_deb:>16.2f}{total_doc_c:>11}{total_cre:>17.2f}")
````
