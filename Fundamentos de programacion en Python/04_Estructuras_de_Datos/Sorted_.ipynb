# Algoritmos de clasificación

Los algoritmos de clasificación son métodos para reorganizar datos automáticamente.

---



## 1. Bubble Sort (Ordenamiento burbuja)

Compara elementos vecinos e intercambia posiciones si están en el orden incorrecto.



### Funcionamiento

1. Recorre la lista
2. Compara pares consecutivos
3. Intercambia si es necesario
4. Repite hasta ordenar completamente


````python
numeros = [5, 2, 8, 1]

for i in range(len(numeros)):
    for j in range(len(numeros)-1):
        if numeros[j] > numeros[j+1]:
            numeros[j], numeros[j+1] = numeros[j+1], numeros[j]

print(numeros)
````


### Analogía

Como reorganizar tareas de un proyecto comparando prioridades una por una.

---



## 2. Selection Sort (Ordenamiento por selección)

Busca el elemento más pequeño y lo coloca en la posición correcta.



### Funcionamiento

1. Encuentra el menor elemento
2. Lo intercambia con el primero
3. Repite con el resto de la lista

### Ejemplo


````python

numeros = [64, 25, 12, 22, 11]

for i in range(len(numeros)):
    minimo = i

    for j in range(i+1, len(numeros)):
        if numeros[j] < numeros[minimo]:
            minimo = j

    numeros[i], numeros[minimo] = numeros[minimo], numeros[i]

print(numeros)

````


### Analogía

Como ordenar libros en una estantería buscando siempre el siguiente libro correcto.

---



## 3. Quick Sort (Ordenamiento rápido)

Divide la lista usando un elemento pivote y ordena cada parte por separado.



### Funcionamiento

1. Selecciona un pivote
2. Divide elementos menores y mayores
3. Ordena recursivamente

### Ejemplo


````python
def quicksort(lista):
    if len(lista) <= 1:
        return lista

    pivote = lista[0]

    menores = [x for x in lista[1:] if x <= pivote]
    mayores = [x for x in lista[1:] if x > pivote]

    return quicksort(menores) + [pivote] + quicksort(mayores)
````

# Clasificación personalizada

Puedes ordenar usando criterios específicos con `key`.



## Ejemplo: ordenar palabras por longitud


````python
palabras = ["python", "sol", "programacion", "ia"]

palabras.sort(key=len)

print(palabras)
````


# Ordenar diccionarios


````python
productos = [
    {"nombre": "Laptop", "precio": 1200},
    {"nombre": "Mouse", "precio": 25},
    {"nombre": "Monitor", "precio": 300}
]

productos.sort(key=lambda x: x["precio"])

print(productos)
````

# Búsqueda lineal

## ¿Qué es?

La búsqueda lineal revisa cada elemento de la lista uno por uno hasta encontrar el objetivo.

---

## Funcionamiento

1. Recorre la lista desde el inicio
2. Compara cada elemento con el objetivo
3. Si encuentra coincidencia, devuelve el índice
4. Si termina la lista sin encontrarlo, devuelve `-1`

---



## Ejemplo de búsqueda lineal


````python
def busqueda_lineal(lista, objetivo):

    for indice, elemento in enumerate(lista):

        if elemento == objetivo:
            return indice

    return -1
````

````python

datos = [10, 20, 30, 40, 50]

resultado = busqueda_lineal(datos, 30)

print(resultado)
````

# Ventajas de la búsqueda lineal

- Fácil de implementar
- Funciona en listas desordenadas
- Buena para listas pequeñas

---

# Desventajas

- Puede ser lenta en listas grandes
- Recorre elementos uno por uno

---


# Cuándo usar búsqueda lineal

Usa búsqueda lineal cuando:

- La lista es pequeña
- La lista no está ordenada
- Hay pocas búsquedas

---



# Búsqueda binaria

## ¿Qué es?

La búsqueda binaria es un algoritmo muy rápido para listas ordenadas.

Divide repetidamente la lista en mitades.

---

# Requisitos

La lista debe estar ordenada previamente.

---

# Funcionamiento

1. Define un rango de búsqueda
2. Encuentra el elemento central
3. Compara con el objetivo
4. Descarta la mitad innecesaria
5. Repite hasta encontrar el elemento

---

# Ejemplo de búsqueda binaria


````python

def busqueda_binaria(lista, objetivo):

    bajo = 0
    alto = len(lista) - 1

    while bajo <= alto:

        medio = (bajo + alto) // 2

        if lista[medio] == objetivo:
            return medio

        elif lista[medio] < objetivo:
            bajo = medio + 1

        else:
            alto = medio - 1

    return -1

````

````python

datos = [10, 20, 30, 40, 50, 60, 70]

resultado = busqueda_binaria(datos, 50)

print(resultado)
````


# Ventajas de la búsqueda binaria

- Extremadamente rápida
- Ideal para listas grandes
- Muy eficiente

---

# Desventajas

- Solo funciona con listas ordenadas
- Más compleja de implementar

---

# Cuándo usar búsqueda binaria

Usa búsqueda binaria cuando:

- La lista es grande
- La lista ya está ordenada
- Realizarás muchas búsquedas
- La velocidad es importante

---



# Comparación entre búsqueda lineal y binaria

| Característica | Búsqueda lineal | Búsqueda binaria |
|---|---|---|
| Requiere orden | No | Sí |
| Velocidad | Más lenta | Muy rápida |
| Implementación | Simple | Más compleja |
| Ideal para | Listas pequeñas | Listas grandes |

# List Comprehension para búsquedas

Permite crear nuevas listas según criterios.


## Ejemplo


````python
numeros = [1, 2, 3, 4, 5, 6]

pares = [n for n in numeros if n % 2 == 0]

print(pares)

````
