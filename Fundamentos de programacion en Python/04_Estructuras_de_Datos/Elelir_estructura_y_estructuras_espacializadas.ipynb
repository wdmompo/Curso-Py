# Estructuras de datos en Python

## Introducción

Las estructuras de datos son fundamentales en Python porque permiten organizar, almacenar y manipular información de manera eficiente.

- Cómo se almacenan los datos

Elegir la estructura correcta puede mejorar:

- El rendimiento
- La velocidad
- La legibilidad
- El mantenimiento del código

---

# ¿Qué es una estructura de datos?

Una estructura de datos es una forma de organizar información en la memoria del computador.

Cada estructura está diseñada para resolver distintos tipos de problemas.

Algunas estructuras son mejores para:

- Mantener orden
- Buscar rápidamente
- Insertar o eliminar elementos
- Manejar grandes cantidades de información

---

# Importancia de elegir la estructura correcta

La elección de una estructura de datos afecta directamente:

- La velocidad del programa
- El uso de memoria
- La claridad del código

# Estructuras de datos comunes en Python

---

# 1. Listas (`list`)

Las listas almacenan elementos ordenados.

## Características

- Mantienen el orden
- Permiten duplicados
- Son mutables

## Ejemplo

```python
frutas = ["manzana", "banana", "pera"]
```

---

## Casos de uso

- Feeds de redes sociales
- Historiales
- Secuencias de datos

---

# 2. Diccionarios (`dict`)

Almacenan pares clave-valor.

## Características

- Acceso rápido
- Claves únicas
- Muy eficientes para búsquedas

## Ejemplo

```python
producto = {
    "id": 101,
    "nombre": "Laptop",
    "precio": 1200
}
```

---

## Casos de uso

- Bases de datos
- Configuraciones
- Sistemas de recomendación

---

# 3. Conjuntos (`set`)

Colecciones de elementos únicos.

## Características

- No permiten duplicados
- Búsquedas rápidas
- No mantienen orden

## Ejemplo

```python
usuarios = {"Ana", "Luis", "Carlos"}
```

---

## Casos de uso

- Inventarios
- Usuarios únicos
- Eliminación de duplicados

---

# Comparación rápida

| Estructura | Orden | Duplicados | Búsqueda rápida |
|---|---|---|---|
| Lista | Sí | Sí | Media |
| Diccionario | Sí* | No en claves | Muy rápida |
| Set | No | No | Muy rápida |

\* Desde Python 3.7 los diccionarios mantienen orden de inserción.

##  Estructuras Especializadas



## deque

Cola de doble extremo.

Uso ideal
Colas
Pilas
Sistemas de deshacer/rehacer


````python
from collections import deque

cola = deque()

cola.append("tarea1")
cola.append("tarea2")

print(cola.popleft())

````

##  heapq



Cola de prioridad.

Uso ideal
Planificación de tareas.
Algoritmos de rutas.
Sistemas de prioridad.


````python
#ejwmplo de headp`q

import heapq

numeros = [5, 2, 9, 1, 5, 6]

heapq.heapify(numeros)
heapq.heappush(numeros, 3)

print(heapq.heappop(numeros))  # Imprime el número más pequeño
````

## collections.Counter



Cuenta ocurrencias de elementos.

Uso ideal
Frecuencia de palabras.
Estadísticas rápidas.


````python
from collections import Counter

texto = "hola hola mundo"

contador = Counter(texto.split())

print(contador)
````

````python
from collections import deque, Counter

# Deque
queue = deque()

queue.append("task1")
queue.append("task2")

print(queue.popleft())

# Counter
text = "This is a sample text with some repeated words words"

counter = Counter(text.split())

print(counter)
````

Compar Velocidad de Busqueda lista vs Dicc

````python
import timeit

# List lookup
list_data = list(range(100000))
lookup_value = 99999
list_time = timeit.timeit(lambda: lookup_value in list_data, number=1000)

# Dictionary lookup
dict_data = {i: i for i in range(100000)}
dict_time = timeit.timeit(lambda: lookup_value in dict_data, number=1000)

print("List lookup time:", list_time)
print("Dictionary lookup time:", dict_time)
````
