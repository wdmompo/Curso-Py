En la celda de abajo, agrega `inventory.csv` dentro de pd.read_csv() para cargar los datos del inventario desde el archivo CSV. Luego, ejecuta la celda, lo cual:
* Extraerá los nombres de los productos del DataFrame y los almacenará en una lista llamada `inventory`.
* Imprimirá la lista de inventario para ver los datos.


```python
# Import a powerful tool called "pandas" that you'll use to work with and organize data easily
import pandas as pd

# Load the inventory from the CSV file
inventory_df = pd.read_csv('inventory.csv')

# Extract the product names from the DataFrame and store them in a list
inventory = inventory_df['product_name'].tolist()

# Print the inventory list and inspect the output
print(inventory) 
```

En la celda a continuación:

Sin embargo, acaba de llegar un nuevo envío de deliciosas golosinas para mascotas.

La variable `new_items_str` en la celda de abajo contiene una cadena separada por comas que representa estos nuevos artículos. El código proporcionado convierte `new_items_str` en una lista de artículos individuales usando el método `split()`.

El método `split()` divide una cadena en una lista de subcadenas basada en un delimitador especificado (en este caso, una coma seguida de un espacio).

En la celda de abajo:
* Usa el método `extend()` para agregar todos los nuevos artículos de la lista `new_items` al final de la lista de inventario existente añadiendo `new_items` al método `extend()`.

Por último, ejecuta la celda para imprimir el inventario actualizado y ver los artículos que has añadido recientemente.


```python
print("A new shipment of gourmet goodies has arrived!")

# New shipment of goodies! (provided as a string)
new_items_str = "Squeaky Sausages, Tuna Tidbits, Crunchy Carrots"

# Convert the string to a list and stores them in the variable new_items
new_items = new_items_str.split(", ") 

print(new_items)

# Update the inventory list by adding "new_items" to the extend() method
inventory.extend(new_items) 

print("\nUpdated inventory:")
print(inventory) 
len(inventory)
```

Las secciones anteriores se centraron en el uso de listas y tuplas para la gestión de inventario. Aunque estas estructuras de datos son útiles, tienen limitaciones cuando se trata de acceder y actualizar de manera eficiente información específica del producto.

Los diccionarios ofrecen un enfoque más organizado y flexible, permitiéndote almacenar nombres de productos como claves y sus niveles de stock correspondientes como valores. Esta estructura clave-valor permite búsquedas y modificaciones rápidas, haciendo que los diccionarios sean ideales para gestionar datos de inventario.

Ejecuta la celda a continuación, que usa la biblioteca pandas para cargar datos desde el archivo CSV y convertirlo en un diccionario. Observa la versatilidad de pandas aquí en el manejo de diferentes estructuras de datos.


```python
import pandas as pd

# Load inventory data from CSV
inventory_df = pd.read_csv('inventory.csv')

# Set the 'product_name' column as the index
inventory_df.set_index('product_name', inplace=True)

# Convert the DataFrame directly to a dictionary, selecting only the 'stock_level' column
inventory_dict = inventory_df['stock_level'].to_dict()

# Print the inventory
print("Initial Inventory:")
print(inventory_dict)
```

Tenga en cuenta que la salida anterior ahora es un diccionario donde los nombres de los productos son las claves, y los niveles de inventario son los valores correspondientes.

Por ejemplo, "Salmon Snacks" es la clave y "25" es el valor.

Esta estructura permite una búsqueda y gestión eficiente de la información del inventario.

Imagine que ha llegado un nuevo envío de productos emocionantes para mascotas a The Furry Foodie y necesita actualizar su diccionario `inventory_dict` para reflejar estos cambios.

En la celda a continuación:

* Agregue un nuevo artículo, "Puppy Snacks", al `inventory_dict` con un nivel de inventario de 40.
* También recibimos inventario adicional de "Cheesy Chompers". Actualice el nivel de inventario de "Cheesy Chompers" en el `inventory_dict` a 20.

Luego, ejecute la celda para imprimir el `inventory_dict` actualizado y ver los cambios.


```python
# Add "Puppy Snacks" to the inventory
inventory_dict["Puppy Snacks"] = 40

# Update "Cheesy Chompers" in the inventory and assign it the value of 20
inventory_dict["Cheesy Chompers"] = 20

print("\nUpdated Inventory:")
print(inventory_dict)
```

¡Observa la salida de arriba! Puppy Snacks con un nivel de stock de 40 ha sido añadido a tu diccionario! El valor del nivel de stock de Cheesy Chompers también ha sido actualizado a 20! Gran trabajo.

Desafortunadamente, escuchas que "Peanut Butter Biscuits" ha sido descontinuado en The Furry Foodie.

Necesitas eliminar este artículo de tu diccionario `inventory_dict` para mantener los registros actualizados.

En la celda de abajo:

* Usa la palabra clave `del` para eliminar la entrada de "Peanut Butter Biscuits" del diccionario `inventory_dict`. Añade `del` al frente de la variable `inventory_dict` y completa con "Peanut Butter Biscuits".

Luego, ejecuta la celda para imprimir el `inventory_dict` actualizado y ver los cambios.


```python
# We've discontinued "Peanut Butter Biscuits". Remove it from the inventory

# Remove item 
del inventory_dict["Peanut Butter Biscuits"] 

print("\nUpdated Inventory after Discontinuation:")
print(inventory_dict)
```

¡Buen trabajo! Ahora, ejecuta la celda de abajo, la cual generará un informe de inventario bien formateado.

Aquí tienes un desglose de la celda final:

* Imprimir un encabezado: Comenzará imprimiendo un encabezado claro "Informe de Inventario:" para separar visualmente el informe de otras salidas.

* Iterar a través del inventario: Usa un bucle `for` para recorrer cada elemento en el diccionario `inventory_dict`.

* Extraer información del producto y stock: Para cada iteración del bucle, se usa el método `items()` del diccionario para obtener tanto el `producto` (clave) como su nivel de `stock` correspondiente (valor).

* Imprimir salida formateada: Usando una cadena f, crea una línea de texto bien formateada para cada producto, mostrando su nombre y nivel de stock actual. Esto crea un informe de inventario claro y organizado.

¡Ejecuta el código y mira el informe!


```python
# It's time to generate a nicely formatted inventory report!

print("\nInventory Report:")
for product, stock in inventory_dict.items(): 
    print(f"Product: {product}, Stock: {stock}") 
```
