# Estructuras de Datos en Python y Aplicaciones del Mundo Real



## Introducción



Una estructura de datos es una forma de organizar información en un ordenador para acceder y utilizar los datos de manera eficiente.

Cada estructura de datos está diseñada para resolver distintos tipos de problemas.

Elegir la estructura adecuada puede mejorar:

- Rendimiento
- Velocidad
- Uso de memoria
- Organización del código

---



# Objetivos



Al estudiar estructuras de datos podrás:

- Comparar diferentes estructuras
- Identificar ventajas y desventajas
- Elegir la estructura correcta para un problema
- Crear aplicaciones más eficientes

---



# Principales Estructuras de Datos



## Estructuras comunes



- Arrays
- Listas enlazadas
- Pilas (Stacks)
- Colas (Queues)
- Árboles (Trees)
- Grafos (Graphs)
- Conjuntos (Sets)
- Diccionarios (Dictionaries)

---



# Arrays



## ¿Qué son?



Los arrays almacenan elementos en posiciones contiguas de memoria.

---

# Características

- Acceso rápido por índice
- Tamaño fijo en muchos lenguajes
- Muy eficientes para lectura

---



# Ejemplo

```python
numeros = [10, 20, 30, 40]
```

---



# Ventajas



- Acceso rápido
- Simples de usar
- Buen rendimiento

---



# Limitaciones



- Inserciones y eliminaciones costosas
- Menos flexibles

---



# Listas Enlazadas



## ¿Qué son?



Son estructuras donde cada elemento apunta al siguiente.

---



# Características

- Elementos conectados mediante punteros
- Inserción eficiente
- Tamaño dinámico

---



# Ventajas

- Flexibles
- Buenas para inserciones frecuentes

---



# Limitaciones

- Acceso lento a elementos
- Mayor uso de memoria

---



# Pilas (Stacks)



## Principio LIFO

```text
Last In, First Out
```

El último elemento agregado es el primero en salir.

---



# Ejemplo Real

## Historial del navegador

Cuando visitas páginas:

- Cada página se agrega a la pila
- El botón “Atrás” elimina la página más reciente

---



# Ejemplo en Python

```python
historial = []

historial.append("google.com")
historial.append("youtube.com")

historial.pop()
```

---



# Colas (Queues)



## Principio FIFO

```text
First In, First Out
```

El primer elemento agregado es el primero en salir.

---

# Ejemplo Real

- Filas de atención
- Procesamiento de tareas
- Impresoras

---



# Ejemplo con deque

```python
from collections import deque

cola = deque()

cola.append("Tarea 1")
cola.append("Tarea 2")

cola.popleft()
```

---



# Árboles (Trees)



## ¿Qué son?

Organizan datos jerárquicamente.

---



# Componentes

- Nodo raíz
- Ramas
- Hojas

---



# Ejemplo Real

## Sistema de archivos

```text
Raíz
├── Documentos
│   └── informe.txt
└── Imágenes
```


# Ventajas



- Excelente organización jerárquica
- Búsquedas eficientes

---


# Uso Común



- Sistemas de archivos
- Bases de datos
- Inteligencia artificial

---



# Grafos (Graphs)



## ¿Qué son?

Representan relaciones entre elementos.

---

# Componentes

- Nodos
- Aristas

---



# Ejemplo Real



## Redes sociales

```python
red_social = {
    "Ana": ["Carlos", "Lucía"],
    "Carlos": ["Ana", "Pedro"]
}
```

---



# Aplicaciones



- Redes sociales

---



# Conjuntos (Sets)



## ¿Qué son?



Colecciones de elementos únicos.

---

# Características

- No permiten duplicados
- Búsqueda rápida
- Pruebas de pertenencia eficientes

---

# Ejemplo

```python
playlist = {
    "Canción A",
    "Canción B",
    "Canción C"
}
```

---

# Verificar Existencia

```python
if "Canción D" in playlist:
    print("Ya existe")
```

---

# Casos de Uso

- Usuarios únicos
- Palabras únicas
- Eliminación de duplicados

---



# Diccionarios

## ¿Qué son?



Estructuras clave-valor.

---

# Ejemplo

```python
usuario = {
    "nombre": "Ana",
    "edad": 25
}
```

---



# Diccionarios Anidados

Permiten organizar información compleja.

---

# Ejemplo

```python
persona = {
    "nombre": "Carlos",
    "contacto": {
        "email": "carlos@email.com",
        "telefono": "123456"
    }
}
```

Las imágenes pueden almacenarse como listas anidadas.

---

# Cómo Elegir la Estructura Correcta

## Preguntas importantes

### ¿Los datos necesitan orden?

- Sí → listas, arrays, colas
- No → sets, diccionarios

---

### ¿Necesitas búsquedas rápidas?

- Sets
- Diccionarios
- Árboles

---

### ¿Necesitas relaciones complejas?

- Grafos
- Diccionarios anidados

---

### ¿Necesitas jerarquías?

- Árboles

---

### ¿Importa mucho el rendimiento?

Analiza:

- Velocidad
- Memoria
- Inserciones
- Eliminaciones
- Acceso aleatorio

---



# Comparación General

| Estructura | Uso Principal | Ventaja |
|---|---|---|
| Array | Acceso rápido | Muy eficiente |
| Lista enlazada | Inserciones frecuentes | Flexible |
| Stack | Historial | Simplicidad |
| Queue | Procesamiento ordenado | FIFO |
| Tree | Datos jerárquicos | Organización |
| Graph | Relaciones complejas | Conexiones |
| Set | Elementos únicos | Búsqueda rápida |
| Dictionary | Clave-valor | Acceso eficiente |

---



# Idea Principal

No existe una estructura de datos perfecta para todos los casos.

La clave es:

- Analizar el problema
- Entender los requisitos
- Elegir la herramienta adecuada

Las estructuras de datos son fundamentales para crear aplicaciones rápidas, organizadas y eficientes en Python.
