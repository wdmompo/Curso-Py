```python
class Banco:
    """
    Clase que representa un banco.

    Attributes:
        name (str): Nombre del titular de la cuenta.
        saldo (float): Saldo de la cuenta.
    """
    
    def __init__(self, name, saldo):

        self.name = name
        self.saldo = saldo
    
    def depositar(self, valor):
        """_summary_

        Args:
            valor (_type_): _description_
        """
        self.saldo += valor
        print(f"Depósito de R${valor} realizado. Novo saldo: R${self.saldo}")
```

```python
emi = Banco("Emi", 1000)
```

## Polimorfismo con Herencia

```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def hablar(self):
        print(f"Hola, mi nombre es {self.nombre} y tengo {self.edad} años.")
```

```python
class Alumno(Persona):
    def __init__(self, nombre, edad, matricula):
        super().__init__(nombre, edad)
        self.matricula = matricula
    
    def hablar(self):
        print(f"Hola, soy un alumno. Mi nombre es {self.nombre}, tengo {self.edad} años y mi matrícula es {self.matricula}.")
```

```python
class Profesor(Persona):
    def __init__(self, nombre, edad, curso):
        # super().__init__(nombre, edad)
        self.nombre = nombre
        self.edad = edad
        self.curso = curso

    def hablar(self):
        print(f"Hola, soy un profesor. Mi nombre es {self.nombre}, tengo {self.edad} años y doy  el curso de {self.curso}.")
```

```python
emi = Profesor("Emi", 30, "python")
ana = Alumno("Ana", 20, "12345")
pedro = Persona("Pedro", 40)
```

```python
for  persona in [emi, ana, pedro]:
    persona.hablar()
```

```python
class Director(Persona):
    def __init__(self, nombre, edad, empresa):
        super().__init__(nombre, edad)
        self.empresa = empresa
```

```python
ariel = Director("Arie", 50, "Tech Company")
```

```python
ariel.hablar()
```

# Polimorfismo sin Herencis Duck Typing

```python
class Sube:
    def __init__(self, nombre, saldo):
        self.name = nombre
        self.saldo = saldo

    def depositar(self, valor):
        self.saldo += valor
        print(f"Depósito de ${valor} a mi tarjeta {self.name}. Saldo actual: ${self.saldo}.")
```

```python
su = Sube("Ariel", 100)
print(su.saldo)
```

```python
santander = Banco("Santander", 5000)
print(santander.saldo)
```

```python
for desde in [su, santander]:
    desde.depositar(1000)
    print(f"Saldo actual en {desde.name}: R${desde.saldo}")
```
