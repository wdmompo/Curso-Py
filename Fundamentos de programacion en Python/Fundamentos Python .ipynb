### Estructura *if*...*elif*...*else*.



Es posible evaluar más de una expresión lógica mediante el uso de _elif_. En el caso de que exista más de una expresión lógica que de por resultado _True_, Python ejecutará solamente el código delimitado por la primera que ocurra.

En caso de que ninguna de las condiciones de por resultado _True_ se puede utilizar _else_ al final de la estructura.

```python
<flujo principal>
...
...
if <expresión lógica>:
     <bloque inscrito al if>
elif <expresión lógica 1>:
     <bloque inscrito al elif>
elif <expresión lógica 2>:
     <bloque inscrito al elif>
...
...
elif <expresión lógica n>:
     <bloque inscrito al elif>
else:
    <bloque inscrito al else>
<flujo principal>
```

## Ciclos con _while_.



Python cuenta con la palabra reservada *while* para ejecutar un bloque de código recursivamente mientras se cumpla una condición determinada. Cuando la expresión lógica evaluada por *while* sea *False* , el flujo de ejecución continuará sin ejecutar el bloque dentro de _while_.

```python
<flujo principal>
...
...
while <expresión lógica>:
     <bloque inscrito a while>
<flujo principal>
```
A continuación se muestra un diagrama de flujo que ejemplifica al uso de *while*:


**Ejemplo:**

````python
i = 1
while i <= 5 :
    # print(i)
    i += 1 # i = i + 1
    print(i)
print("continua el flujo")
````

## Romper un bucle con break

````python

while True:
    pre = input("terminar ?")
    if pre == "s":
        print(pre)
        break
    else:
        print("sigue")
````

````python
terminar = True
while terminar:
    pre = input("terminar ?")
    if pre == "s":
        terminar = False
    else:
        print("sigue")
````

La forma más común de realizar iteraciones en otros lenguajes de programación es por medio algo similar al uso de la función *range().*

### La función *range()*

Para definir rangos numéricos se usa la función *range()*.

* *range(*n, m, s*)* cumple:  rango >= n and rango < m en incrementos de s.
* *range(*n, m*)* cumple:  rango >= n and rango < m en incrementos de 1.
* *range(*m*)* cumple:  rango >= 0 and rango < m en incrementos de 1.

**Ejemplos:**

````python
range(10)
````

````python
for i in range(11):
    print(i)
````

````python
for i in range(-10,11):
    print(i)
````

````python
for i in range(-10,11,2):
    print(i)
````

````python
for i in range(10,-11,-2):
    print(i)
````

````python
for año in range(1930,2025):
    if año % 4 == 0 and año % 100 != 0 or año % 400 == 0:
        print(año,"es bisiesto")
````

## Saltos de iteraciones con **continue**

````python
for año in range(1930,2023,4):
    print(año)
    if año != 1978 and año != 1986 and año != 2022:
        continue
    print(año,"ARG CAMPEON FIFA")
````

````python
def cenamos (lista): 
    """
    Función que determina qué cenar, qué beber y qué postre disfrutar, 
    dependiendo de los elementos pasados en una lista como argumento.

    Parámetros:
    lista (list): Lista que contiene 2 elementos 
                 [ cena , postre ] postre opcional 

    Returns:
    str: Mensaje indicando la cena, bebida y postre seleccionados.
    """
    match lista:
        
        case [("Pizza" | "Empanadas"  as cena ), postre]:
            bebida = "Cerveza"
        
        case [("Asado" | "Fideos" as cena), postre]:
            bebida = "Vino"
        
        case [cena, postre]:
            bebida = "Agua"
        
        case [cena]:
            bebida = "Agua"
            postre = "Nada"
        
        case [] :
            return "Hoy no se cena"
        
        case _ :
            raise TypeError (f"Error necesita pasar 2 argumenos como max y se recibieron {len(lista)}") 
            #return f"Error necesita pasar 2 argumenos como max y se recibieron {len(lista)}"   
    
    return f"Vamos a cenar {cena} , tomamos {bebida} y de postre {postre}"
````

### Captura de varios argumentos en un parámetro de tipo *dict* (kargs).

````python
def superficie(**dato):
    '''Calcula la superficie de una figura geométrica si los parámetros  ingresados
       coinciden.'''
    print(type(dato))
    print(dato)
    
    if dato["tipo"] == "Rectángulo":
        superficie = float(dato["base"]) * float(dato["altura"])
    elif dato["tipo"] == "Triángulo":
        superficie = float(dato["base"]) * float(dato["altura"]) / 2
    elif dato["tipo"] == "Círculo":
        superficie = float(dato["radio"]) ** 2 * 3.14259265
    else:
        print("No puedo calcular la superficie.")
    print("La superficie del %s es de %.3f" % (dato["tipo"].lower(), superficie))
````

````python
superficie(base=22, altura=30, tipo="Rectángulo")
````

````python
superficie(tipo="Círculo", radio = 35)
````

````python
def factorial(numero):
    print('En este ámbito, numero =', numero)
    if numero == 1:
        print('Llegó al final.\nRegresa 1!')
        return 1
    else:
        fact = numero * factorial(numero - 1)
        print('Regresa %d!: %d' %(numero, fact))
        return fact
````

````python
def factorial(numero):
    """Calcula el factorial de un número dado.

    Args:
        numero: El número del que se quiere calcular el factorial.

    Returns:
        El factorial del número dado.

    Raises:
        TypeError: Si el número no es un entero.
        ValueError: Si el número es negativo.
    """

    if not isinstance(numero, int):
        raise TypeError("El número debe ser un entero.")

    if numero < 0:
        raise ValueError("El número debe ser positivo.")

    if numero == 1:
        return 1
    else:
        fact = numero * factorial(numero - 1)
        return fact
````
