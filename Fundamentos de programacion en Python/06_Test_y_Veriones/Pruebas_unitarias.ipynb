# Pruebas Unitarias y pytest
**Objetivo:** Cobertura compacta y completa de conceptos clave para aprender y aplicar pruebas unitarias en Python con `pytest`.

## 1) ¿Qué son las pruebas unitarias? (Resumen)
- Verifican el comportamiento de unidades aisladas (funciones/métodos).
- Detectan errores tempranos, facilitan refactorizaciones y actúan como documentación viva.
- Ciclo: escribir código → escribir pruebas → ejecutar → corregir → repetir.

## 2) `pytest` — ¿por qué usarlo?
- Sintaxis simple, descubrimiento automático de pruebas (`test_`), fixtures y parametrización integradas.
- Instalación: `pip install pytest` (ejecución: `pytest`).

```python
# Ejemplo: funciones pequeñas a probar
def add(a, b):
    return a + b

def calculate_discount(price, percentage):
    return price - (price * percentage / 100)

# Caso: comportamiento esperado
if __name__ == '__main__':
    print('add(2,3)=', add(2,3))
    print('discount 100,10% =', calculate_discount(100,10))
```

## 3) Escribir pruebas con `pytest` — ejemplos rápidos
- Las pruebas son funciones que comienzan con `test_` y usan `assert`.
- Se pueden ejecutar con `pytest -q` para salida compacta.

```python
# test_examples.py — ejemplos de pruebas (no ejecutar aquí, ejecutar con pytest)
import pytest

def test_add():
    assert add(2, 3) == 5

def test_calculate_discount():
    assert calculate_discount(100, 10) == 90
```

## 4) Fixtures en `pytest`
- Permiten preparar/limpiar recursos reutilizables con `@pytest.fixture`.
- Scopes: `function` (por defecto), `class`, `module`, `session` — elegir según coste de setup.
- Útiles para datos de prueba, clientes de HTTP, conexiones a DB, etc.

```python
# Ejemplo de fixture y uso
import pytest

@pytest.fixture
def sample_data():
    return [1, 2, 3]

def test_sum(sample_data):
    assert sum(sample_data) == 6

@pytest.fixture(scope='module')
def expensive_setup():
    # Simula inicialización costosa
    obj = {'conn': 'db_connection'}
    yield obj
    # teardown aquí
    obj.clear()
```

## 5) Parametrización y marcadores
- `@pytest.mark.parametrize` para probar múltiples entradas con una sola función.
- `@pytest.mark.skip` / `@pytest.mark.xfail` y marcadores personalizados para organizar ejecuciones.

```python
# Ejemplo de parametrización
import pytest

@pytest.mark.parametrize('a,b,expected', [(2,3,5),(0,0,0),(-1,1,0)])
def test_add_param(a,b,expected):
    assert add(a,b) == expected

# Uso de marcador
@pytest.mark.skip(reason='demostración')
def test_skip_example():
    assert False
```

## 6) Organización de pruebas
- Estructura recomendada: `project/` + `tests/` con archivos `test_*.py`.
- `conftest.py` para fixtures compartidos y configuración global.
- Mantener una prueba = una responsabilidad; nombres descriptivos.

```python
# conftest.py — ejemplo breve de fixtures compartidos
import pytest

@pytest.fixture
def test_product():
    return {'name': 'Laptop', 'price': 1000}

@pytest.fixture
def empty_cart():
    return []
```

## 7) Buenas prácticas (Checklist rápida)
- Escribir tests antes o junto al código al refactorizar.
- Mantener tests independientes y reproducibles.
- Usar fixtures para evitar duplicación.
- Parametrizar casos similares.
- Ejecutar tests en CI (automatizar).
- Revisar y actualizar tests cuando cambie la API.
