### Actividad 1: Tomando decisiones con instrucciones condicionales
#### Introducción
Como analista de datos para una librería en línea, se te encarga analizar los patrones de compra de los clientes. El equipo de marketing quiere enviar códigos de descuento especiales a los clientes según la cantidad de libros que hayan comprado. Si un cliente ha comprado suficientes libros, recibirá un descuento. Si ha comprado más, podría calificar para un descuento aún mejor.
#### Tarea 1: Definir una función de descuento
Necesitas escribir una función que ayude al equipo de marketing a decidir qué clientes deben recibir un descuento.

Define una función llamada ```send_discount``` que acepte dos argumentos:
- ```books_purchased```: La cantidad de libros que un cliente ha comprado.
- ```discount_threshold```: La cantidad mínima de libros que un cliente necesita comprar para recibir un descuento.

La función debe imprimir uno de los dos mensajes posibles, como se muestra a continuación. Ten cuidado de no omitir la puntuación en tu declaración de impresión.
- ```Discount applied!``` si el cliente califica para un descuento.
- ```No discount.``` si el cliente no califica.

#### Ejemplo:

```python
send_discount(books_purchased=3, discount_threshold=5)
# Salida: No discount.

send_discount(books_purchased=7, discount_threshold=5)
# Salida: Discount applied!

```

#### Plantilla de código:


````python
def send_discount(books_purchased, discount_threshold):
    if books_purchased >= discount_threshold:
        print("Discount applied!")
    else:
        print("No discount.")
````

### Actividad 2: Uso de bucles para tareas repetitivas
#### Introducción
Eres un gerente de producto en una startup de tecnología. Los clientes han estado proporcionando calificaciones para la última aplicación de la empresa. El equipo de marketing quiere categorizar estos comentarios en tres categorías: "Bajo", "Medio" y "Alto" calificaciones. Utilizarás técnicas de iteración de Python para procesar las calificaciones de los clientes de manera eficiente.
#### Tarea 1: Categorizar calificaciones de clientes
Define una función llamada `categorize_ratings` que toma una lista de calificaciones de clientes como entrada. Cada calificación es un número entero entre 1 y 10.

Tu función categorizará las calificaciones como:
- **Bajo** (1-4)
- **Medio** (5-7)
- **Alto** (8-10)

La salida debe imprimir tres declaraciones, una para cada categoría, en el siguiente orden (Bajo, luego Medio, luego Alto):
<br>`Bajo: {número_de_calificaciones_bajo}`<br>
`Medio: {número_de_calificaciones_medio}`<br>
`Alto: {número_de_calificaciones_alto}`<br>

#### Ejemplo:
```python
# Hay dos calificaciones en el rango 1-4, dos calificaciones en el rango de 5-7 y dos calificaciones en el rango 8-10
categorize_ratings([1, 3, 5, 7, 8, 9])
```
**Salida:**
<br>`Bajo: 2`<br>
`Medio: 2`<br>
`Alto: 2`<br>
#### Plantilla de código:


````python
def categorize_ratings(rating_list):
    # YOUR CODE HERE
    categorized_ratings = {"high": 0, "medium": 0, "low": 0}
    for rating in rating_list:
        if rating >= 8:
            categorized_ratings["high"] += 1
        elif rating >= 5:
            categorized_ratings["medium"] += 1
        else:
            categorized_ratings["low"] += 1
    print(f"Low :{categorized_ratings['low']}\nMedium: {categorized_ratings['medium']}\nHigh: {categorized_ratings['high']}")

    return categorized_ratings
````

#### Task 1: Creating and Sorting Test Scores
Tienes una lista de estudiantes y sus puntajes de exámenes correspondientes. Tu tarea es organizar y analizar los puntajes usando algoritmos de ordenamiento y mecanismos de manejo de errores de Python.
#### Step 1: Create the List of Students
Crea una lista llamada `students` que contenga los siguientes nombres de estudiantes: John, Lisa, Mary, Chris, Linda, Matt


````python
### YOUR CODE HERE ###
students = str.split("John, Lisa, Mary, Chris, Linda, Matt", ", ")
students
````

#### Paso 2: Crear un diccionario de puntajes de prueba
Crear un diccionario llamado `test_performance` y asignar las siguientes puntuaciones a cada estudiante de la siguiente manera:
- John: 87
- Lisa: 90
- Mary: 75
- Chris: 100
- Linda: 100
- Matt: 70


````python
### YOUR CODE HERE ###
test_performance = {
    "John": 87, 
    "Lisa": 90, 
    "Mary": 75, 
    "Chris": 100, 
    "Linda": 100, 
    "Matt": 70
    }

test_performance
````

#### Paso 3: Extraer las puntuaciones del diccionario
Crear una lista llamada `scores` y extraer la puntuación de cada estudiante usando un bucle `for`.


````python
### YOUR CODE HERE ###
# scores = list(test_performance.values())

scores = []
for student in students:
    scores.append(test_performance.get(student, 0))

scores
````

#### Paso 4: Ordenar las puntuaciones con una función personalizada
Define una función llamada `bubble_sort` que ordena la lista de puntuaciones en orden ascendente.


````python
### YOUR CODE HERE ###
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
````

#### Paso 5: Asignar las puntuaciones ordenadas a `sorted_scores`
Llamar a la función `bubble_sort` que definiste arriba y asignar el valor de retorno a `sorted_scores`.


````python
### YOUR CODE HERE ###
sorted_scores = bubble_sort(scores)
sorted_scores
````

Define una función llamada `average_class_score` para calcular el puntaje promedio. Agrega manejo de errores para los casos en que la lista de estudiantes esté vacía.


````python
### YOUR CODE HERE ###
def average_class_score(students, student_scores):
    try:
        total_score = sum(student_scores)
        average_score = total_score / len(student_scores)
        return average_score
    except ZeroDivisionError:
        return 0
````
