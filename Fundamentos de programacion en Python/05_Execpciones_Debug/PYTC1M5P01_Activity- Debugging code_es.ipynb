#### **El TypeError - Enteros**

A continuación, explorarás cómo acceder a información específica dentro de tus diccionarios de insectos. Intentarás recuperar el primer dígito del número de patas que tiene un insecto. Recuerda, los diccionarios usan claves para acceder a sus valores.

***Primero, ejecuta la celda de abajo y revisa el error.***

##### **Consejos para Depuración**

* **Entendiendo el Error:** El `TypeError` indica que estás intentando usar el operador de subíndice [] (utilizado para acceder a elementos dentro de secuencias como listas o cadenas) en un entero, lo cual no está permitido.
* Verifica Tipos de Datos: Usa la función `type()` para confirmar el tipo de dato de una variable si no estás seguro.
* Piensa Lógicamente: Los enteros representan valores numéricos únicos, por lo que intentar acceder a un "elemento" dentro de ellos no tiene sentido.
* Revisa Tu Código: Examina cuidadosamente la línea que causa el error y considera qué estás intentando lograr. ¿Estás usando el enfoque correcto para el tipo de dato con el que estás trabajando?

##### **¿Aún Atascado?**
* Si aún estás atascado, elimina o comenta el código bajo `# Error deliberado` y descomenta el código debajo de la siguiente línea para resolver el error: `# Enfoque correcto (comentado para referencia)`.
* Consejo útil: Puedes comentar y descomentar varias líneas de código a la vez seleccionándolas y presionando `Ctrl + /` en tu teclado.


```python
# Deliberate Error: Trying to access an element within an integer
first_insect = all_insects_data[0]
# first_digit_of_legs = first_insect['legs'][0] 

# Correct approach (commented out for reference)

try :
    first_digit_of_legs = first_insect['legs'][0] 
except TypeError as e:
    print(f"An error occurred: {e}")
    first_digit_of_legs = str(first_insect['legs'])[0]

# first_digit_of_legs = str(first_insect['legs'])[0] 

print(f"The first digit of the number of legs the {first_insect['name']} has is: {first_digit_of_legs}")
```

##### **Por qué funciona el enfoque correcto**

El enfoque corregido funciona porque aborda el problema principal del TypeError. Aquí hay un desglose:

* Conversión de tipos: Usamos str(first_insect['legs']) para convertir el valor entero que representa el número de patas en una cadena. Esto es crucial porque las cadenas son secuencias de caracteres, lo que nos permite acceder a caracteres individuales mediante indexación.

* Indexación: Una vez que tenemos una representación en cadena del número de patas, podemos usar [0] para acceder a su primer carácter, que corresponde al primer dígito del número original.

En esencia, estamos aprovechando la naturaleza similar a una secuencia de las cadenas para extraer la información deseada de un valor numérico.

Esta técnica es útil siempre que necesites manipular o analizar dígitos individuales dentro de un número. Recuerda, entender los tipos de datos y sus capacidades es clave para una depuración y resolución de problemas efectiva en Python.

#### Agregar Nueva Información - El AttributeError

A continuación, intentarás agregar el color 'marrón' a tu diccionario `first_insect`. Intentarás usar erróneamente el método `.append()`, que está diseñado para listas, no para diccionarios.

***Primero, ejecuta la celda de abajo y revisa el error.***

##### **Consejos para Depuración**

* **Entiende el Error:** El `AttributeError` dice que estás intentando usar un método (`.append()`) que no existe para el tipo de dato diccionario
* Conoce Tus Estructuras de Datos:
    * Los diccionarios usan pares clave-valor. Para agregar un nuevo par clave-valor, simplemente asigna un valor a una nueva clave: `dictionary['nueva_clave'] = nuevo_valor`
    * Las listas son colecciones ordenadas de elementos. Para agregar un elemento al final de una lista, usa el método `.append()`: `lista.append(nuevo_elemento)`
    * Revisa la Documentación: Si no estás seguro sobre los métodos disponibles para un tipo de dato en particular, consulta la documentación oficial de Python
    
##### **¿Aún Atascado?**

* Asignación de clave-valor: En los diccionarios de Python, puedes agregar un nuevo par clave-valor o actualizar uno existente usando el operador de asignación `=`. La sintaxis es `diccionario[clave] = valor`.

* Naturaleza dinámica de los diccionarios: Los diccionarios son mutables, lo que significa que puedes modificarlos agregando, actualizando o eliminando pares clave-valor después de que hayan sido creados.
