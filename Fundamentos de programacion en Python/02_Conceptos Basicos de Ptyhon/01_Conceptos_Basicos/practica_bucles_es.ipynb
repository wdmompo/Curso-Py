### Escenario: Domando datos de temperatura con Python
Eres un analista de datos trabajando en un proyecto de clima. Has recopilado una serie de lecturas de temperatura en Celsius,
pero tus herramientas de análisis e informes requieren temperaturas en Fahrenheit.
¡Es hora de poner en práctica tus habilidades en Python!
Crearás un programa que convierte de manera eficiente una lista de temperaturas en Celsius en sus equivalentes en Fahrenheit,
demostrando el poder de los bucles y variables en el manejo de transformaciones de datos del mundo real.

### Resumen de la actividad:
- Crear e inicializar listas para almacenar datos de temperatura.
- Usar un bucle para iterar a través de temperaturas en Celsius.
- Aplicar la fórmula de conversión de Celsius a Fahrenheit.
- Almacenar las temperaturas convertidas en Fahrenheit en una nueva lista.


```python
# Step 1: Store temperature data
celsius_temperatures = [0, 10, 25, 32, 100]  # Celsius temperatures list
fahrenheit_temperatures = []  # Empty list for Fahrenheit temperatures

# Printing initial state of lists
print("Celsius Temperatures:", celsius_temperatures)
print("Fahrenheit Temperatures:", fahrenheit_temperatures)
```

### Paso 2: Convertir temperaturas
Es hora de convertir tus temperaturas en Celsius a Fahrenheit:
- Se ha proporcionado un bucle for para iterar a través de cada valor en Celsius en la lista celsius_temperatures.
- Dentro del bucle:
  - Aplicar la fórmula de conversión de Celsius a Fahrenheit: F = (C * 9/5) + 32, donde F es la temperatura en Fahrenheit y C es la temperatura en Celsius.
  - El código para almacenar cada valor de Fahrenheit calculado en la lista fahrenheit_temperatures usando el método append() ha sido proporcionado para ti.
- Imprimir los resultados


```python
# Lists from Step 1
celsius_temperatures = [0, 10, 25, 32, 100]
fahrenheit_temperatures = []

# Convert each Celsius temperature to Fahrenheit
for celsius in celsius_temperatures:  #Start the loop
    fahrenheit = (celsius * 9/5) + 32  # Celsius to Fahrenheit conversion formula
    fahrenheit_temperatures.append(fahrenheit)  # Append to the list

# Print the results (including the output from Step 1)
print("Celsius Temperatures:", celsius_temperatures)
print("Fahrenheit Temperatures:", fahrenheit_temperatures)
```
