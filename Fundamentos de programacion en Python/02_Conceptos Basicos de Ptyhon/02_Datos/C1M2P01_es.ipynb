### Por qué las listas son perfectas para listas de compras:

* **Colección ordenada:** Las listas mantienen el orden en que agregas los artículos, lo cual puede ser útil para priorizar tu viaje de compras.
* **Mutables:** Puedes agregar, eliminar o modificar artículos fácilmente a medida que cambian tus necesidades.
* **Flexibles:** Las listas pueden almacenar artículos de diferentes tipos (por ejemplo, cadenas para nombres de artículos, números para cantidades).

Para esta actividad, tendrás acceso a un archivo CSV llamado `grocery_list.csv` que ya ha sido creado para ti y contiene algunos artículos iniciales para nuestra lista de compras.

Comienza ejecutando la celda a continuación, que importa la biblioteca pandas, carga el archivo csv llamado `grocery_list.csv`, y luego imprime la salida.


```python
# Import a powerful tool called "pandas" that you'll use to work with and organize data easily
import pandas as pd
import os 

file_name = 'grocery_list.csv'

if os.path.exists(file_name):
    # Load the grocery list from the CSV file
    grocery_list_df = pd.read_csv(file_name)

    # Extract the items from the DataFrame and store them in a list
    grocery_list = grocery_list_df['item'].tolist()

    # Print the grocery list and inspect the output
    print(grocery_list)
else:
    print(f"Error: The file '{file_name}' was not found in the current directory.")
    print("Please ensure you have followed the instructions on the Coursera site to extract the files from the zip archive.")
```

```python
grocery_list.remove("leche")
```
