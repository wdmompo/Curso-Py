```python


def format_output (func):
    def inner (i, j):
        print ('{:3}|'.format (func (i, j)), end = '')
    return inner

@format_output
def product (i, j):
    return (i * j)

@format_output
def pow (i, j):
    return (i ** j)

print ('Table of products')
for i in range (1, 11):
    for j in range (1, 11):
        product (i, j)
    print ('\n' + 10 * '---+')

print ('\nTable of powers')
for i in range (1, 6):
    for j in range (1, 5):
        pow (i, j)
    print ('\n' + 4 * '---+')
```
