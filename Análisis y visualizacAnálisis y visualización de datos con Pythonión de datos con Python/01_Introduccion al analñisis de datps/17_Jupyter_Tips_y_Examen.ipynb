## 1) Que es Jupyter Notebook

Jupyter es un entorno interactivo donde combinas:
- Codigo (Python)
- Texto (Markdown)
- Resultados inmediatos
- Visualizaciones

La unidad de trabajo es la **celda**.

## 2) Tipos de celdas que debes dominar

### Celda de codigo
Ejecuta instrucciones Python.

### Celda Markdown
Sirve para explicar, documentar, crear titulos, listas y formulas.

Esto es importante para examen: **las celdas de codigo y Markdown son caracteristicas interactivas del notebook**.

```python
# Ejemplo rapido de celda de codigo
x = 10
y = 25
suma = x + y
suma
```

## 3) Atajos de teclado clave (muy preguntables)

En **modo comando** (pulsa `Esc`):
- `B`: crea celda abajo
- `A`: crea celda arriba
- `DD`: elimina celda
- `Y`: convierte a codigo
- `M`: convierte a Markdown

Pregunta tipica de examen:
- Formas validas de agregar celda nueva: **A y B**

## 4) Comandos magicos mas utiles

Los comandos magicos (empiezan con `%`) mejoran la ejecucion y analisis.

- `%whos`: lista variables definidas (con tipo e info)
- `%pwd`: muestra directorio actual
- `%ls`: lista archivos y carpetas
- `%timeit`: mide tiempo de ejecucion

```python
# Variables de ejemplo para usar %whos
curso = 'Analisis de datos'
alumnos = 30
promedios = [8.5, 9.0, 7.8]
%whos
```

```python
# Directorio actual y archivos visibles
%pwd
%ls
```

```python
# Medicion de tiempo
%timeit sum(range(10000))
```

## 5) Funciones interactivas: widgets, Markdown y codigo

Otra idea importante de examen: en Jupyter puedes usar:
- Celdas de codigo
- Celdas Markdown
- Widgets interactivos (si esta disponible `ipywidgets`)

```python
# Demostracion opcional de widget
# Si ipywidgets no esta instalado, no pasa nada: se muestra un mensaje amigable.
try:
    import ipywidgets as widgets
    from IPython.display import display

    slider = widgets.IntSlider(
        value=5,
        min=0,
        max=10,
        step=1,
        description='Nivel:'
    )
    display(slider)
except Exception as e:
    print('No se pudo cargar ipywidgets. Puedes instalarlo si lo necesitas.')
    print('Detalle:', e)
```
