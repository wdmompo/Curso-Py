# Archivos CSV: Almacenando Datos en un Formato Tabular

CSV (Valores Separados por Comas) es un formato de archivo común para almacenar datos tabulares, como los que podrías ver en una hoja de cálculo. Es un formato simple y ampliamente soportado, lo que lo convierte en una excelente opción para compartir e intercambiar datos.

En un archivo CSV, cada fila representa un registro u observación, y cada columna representa un atributo o característica específica. Los valores dentro de cada fila están separados por comas (de ahí el nombre).


# Leer archivos CSV en Jupyter Notebooks

Una de las características más poderosas de los Jupyter Notebooks es su capacidad para integrarse sin problemas con fuentes de datos externas. Los archivos CSV, siendo un formato común y versátil, son un ejemplo principal. Podemos usar la biblioteca `pandas` para leer archivos CSV sin esfuerzo directamente en nuestro cuaderno, transformando datos en bruto en un DataFrame estructurado que está listo para su análisis.

En la celda a continuación, leerás el archivo `planets_with_moons_all.csv`, que contiene información sobre los planetas en nuestro sistema solar, incluyendo sus diámetros y el número de lunas que tienen.

El código para importar `pandas` y para imprimir los resultados del csv ha sido proporcionado para ti. Sin embargo, necesitas agregar `planets_with_moons_all.csv` en `pd.read_csv` y luego ejecutar la celda.


```python
import pandas as pd

# Read the CSV file into a DataFrame
planets_df = pd.read_csv('planets_with_moons_all.csv')

# Display the first 5 rows
print(planets_df.head())

# Print information about the DataFrame
print(planets_df.info())
```

# Combinando datos CSV con `numpy` para análisis

Ahora que has cargado datos desde un archivo CSV, también puedes usar bibliotecas externas como `numpy` para realizar cálculos -- todo dentro del mismo entorno de Jupyter Notebook! Es un testimonio de la flexibilidad y el poder que ofrecen los Jupyter Notebooks para flujos de trabajo de análisis de datos.

El código para calcular algunas estadísticas sobre los planetas en nuestro sistema solar ya ha sido proporcionado para ti.

¡Adelante y ejecútalo para ver los resultados!


```python
import numpy as np

# Extract the 'Diameter (km)' and 'Number of Moons' columns as numpy arrays
diameters = planets_df['Diameter (km)'].values
num_moons = planets_df['Number of Moons'].values

# Calculate the mean and median
mean_diameter = np.mean(diameters)
median_diameter = np.median(diameters)
mean_moons = np.mean(num_moons)
median_moons = np.median(num_moons)

# Display the results
print("\nDescriptive Statistics:")
print(f"Mean Diameter: {mean_diameter:.2f} km")
print(f"Median Diameter: {median_diameter} km")
print(f"Mean Number of Moons: {mean_moons:.2f}")
print(f"Median Number of Moons: {median_moons}\n")
```

# Visualización con Seaborn

Jupyter Notebooks no son solo para cálculos; también son fantásticos para crear visualizaciones que nos ayudan a entender mejor nuestros datos. En la siguiente celda, usarás la biblioteca `seaborn` para crear un diagrama de dispersión simple para visualizar la relación entre el diámetro de un planeta y el número de lunas que tiene.

El código necesario para crear esta visualización ha sido proporcionado para ti.

¡Todo lo que necesitas hacer es agregar `plt.show()` como la última línea debajo de `# Mostrar el gráfico` y ejecutar la celda!


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the scatter plot
sns.scatterplot(x='Diameter (km)', y='Number of Moons', data=planets_df)

# Add labels and a title
plt.title('Relationship between Diameter and Number of Moons')
plt.xlabel('Diameter (km)')
plt.ylabel('Number of Moons')

# Display the plot
plt.show()
```
