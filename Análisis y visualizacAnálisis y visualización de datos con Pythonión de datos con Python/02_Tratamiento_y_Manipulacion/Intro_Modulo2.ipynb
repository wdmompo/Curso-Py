## Acceso a elementos at / iat / loc / iloc

### Diferencia entre `at` / `iat` y `loc` / `iloc`

| Método | Base | Qué devuelve | ¿Permite rangos o varios elementos? | Ejemplo típico |
|---|---|---|---|---|
| `at` | etiquetas (label-based) | una sola celda | No | `df.at[0, 'Producto']` |
| `iat` | posición entera | una sola celda | No | `df.iat[0, 0]` |
| `loc` | etiquetas | una fila, una columna, o un subconjunto | Sí | `df.loc[0:2, ['Producto', 'Ventas']]` |
| `iloc` | posición entera | una fila, una columna, o un subconjunto | Sí | `df.iloc[0:2, 0:2]` |

#### Idea clave
- `at` / `iat` están pensados para acceso muy rápido a un único valor.
- `loc` / `iloc` son más flexibles y sirven para filtrar, cortar y seleccionar porciones del DataFrame.

````python
ventas.at[0, 'Producto'] = 'Portátil'

display(ventas)
````

## Ejemplo 3: valores faltantes y limpieza


### Breve explicación sobre el tratamiento de valores faltantes en pandas

En el análisis de datos, es común encontrar valores faltantes (NaN). Pandas ofrece varias formas de tratarlos:

- `fillna()`: Rellena los valores faltantes con un valor específico
- `dropna()`: Elimina filas o columnas que contienen valores faltantes.
- `ffill()`: Propaga el último valor válido hacia adelante para rellenar NaN.
- `bfill()`: Propaga el siguiente valor válido hacia atrás para rellenar NaN.
- `interpolate()`: Rellena NaN usando interpolación (por ejemplo, lineal).

Ejemplo:

````python
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [np.nan, 2, 3, 4]
})

display(df)
````

````python
df.info()
````

````python
df.describe()
````

````python
# Rellenar valores faltantes con un valor fijo
df_fill = df.fillna(0)
display(df_fill)
````

````python
# Rellenar valores faltantes por tipo de columna 
df_fill_x_valor = df.fillna({"A":0, 'B':5})
display(df_fill_x_valor)
````

````python
# Eliminar filas con cualquier valor faltante
display(df)
df_drop = df.dropna()
display(df_drop)
````

````python
# Rellenar valores faltantes propagando el último valor válido hacia adelante
df_ffill = df.ffill()
display(df)
df_ffill
````

````python
# Rellenar valores faltantes propagando el siguiente valor válido hacia atrás
df_bfill = df.bfill()
display(df)
df_bfill
````

````python
# Rellenar valores faltantes usando interpolación lineal
df_interpolate = df.interpolate()
display(df)
df_interpolate
````

## Ejemplo 4: agrupación y agregación


En pandas, la función groupby permite agrupar un DataFrame según una o más columnas, para luego aplicar funciones de agregación o transformación sobre cada grupo. Esto es muy útil para resumir datos, calcular estadísticas por categorías o segmentar información.

La agregación se realiza con métodos como agg o aggregate, que permiten aplicar una o varias funciones a las columnas agrupadas, devolviendo un resumen compacto.

- groupby: Agrupa el DataFrame por una o varias columnas.
- agg o aggregate: Aplica funciones de agregación (como suma, media, conteo, etc.) a los grupos creados.

````python
ventas
````

````python
ventas_region = ventas.groupby('Region')['Ventas'].sum().reset_index()
print('Ventas por región:\n')
ventas_region
````

````python
promedio_region = ventas.groupby('Region')['Ventas'].mean().reset_index()
print('\nPromedio por región:\n')
promedio_region
````

````python
data = {
    'Categoria': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Ventas': [100, 150, 200, 250, 300, 350],
    'Cantidad': [1, 2, 3, 4, 5, 8]
}
data_df = pd.DataFrame(data)

display(data_df)
````

Agrupar por categoría y calcular suma de ventas y cantidad

````python
resultado = data_df.groupby('Categoria').agg({
    'Ventas': 'sum',
    'Cantidad': 'sum'
})
resultado
````

Aplicar múltiples funciones a una columna

````python
resultado_multi = data_df.groupby('Categoria').agg({
    'Ventas': ['sum', 'mean', 'max'],
    'Cantidad': 'mean'
})
resultado_multi
````

Usando Funciones Propias

````python
# Definir una función propia que calcule el rango (max - min)
def rango(x):
    return x.max() - x.min()
````

````python
# Aplicar groupby con la función propia usando agg
resultado = data_df.groupby('Categoria').agg({
    'Ventas': rango,
    'Cantidad': rango
})

resultado
````

## Ejemplo 5: merge y concat


### Merge en pandas

Se usa para combinar dos DataFrames basándose en una o más columnas clave, similar a una unión (join) en bases de datos. Permite hacer combinaciones tipo 

- inner
- left
- right
- outer.

Ideal para combinar datos relacionados con claves comunes.

### Concat en pandas

Se usa para concatenar (unir) DataFrames a lo largo de un eje (filas o columnas).
Útil para apilar DataFrames uno encima del otro (por filas) o lado a lado (por columnas).
No requiere columnas clave, simplemente une los datos.

Creamos dos tablas 

````python
clientes = pd.DataFrame({
    'id': [1, 2, 3],
    'Cliente': ['Ana', 'Luis', 'Mara']
})
clientes
````

````python
compras = pd.DataFrame({
    'id': [1, 2, 4],
    'Compra': [100, 200, 300]
})
compras
````

#### Explicación del método merge con las opciones de 'how' en pandas

El método `merge` de pandas se utiliza para combinar dos DataFrames en base a una o más columnas clave.

Parámetro clave: `how`
Define el tipo de unión que se realizará entre los DataFrames. Las opciones principales son:

- 'left': Realiza una unión izquierda. Se mantienen todas las filas del DataFrame izquierdo y se agregan las filas coincidentes del DataFrame derecho. Si no hay coincidencia, se rellenan con NaN.
- 'right': Realiza una unión derecha. Se mantienen todas las filas del DataFrame derecho y se agregan las filas coincidentes del DataFrame izquierdo. Si no hay coincidencia, se rellenan con NaN.
- 'inner': Realiza una unión interna. Solo se mantienen las filas que tienen coincidencias en ambos DataFrames.
- 'outer': Realiza una unión externa completa. Se mantienen todas las filas de ambos DataFrames, rellenando con NaN donde no haya coincidencias.

````python
unido_iner = pd.merge(clientes, compras, on='id', how='inner')
print('Merge:\n')
unido_iner
````

````python
unido_left = pd.merge(clientes, compras, on='id', how='left')
print('Merge:\n')
unido_left
````

````python
unido_right = pd.merge(clientes, compras, on='id', how='right')
print('Merge:\n')
unido_right
````

````python
unido_outer= pd.merge(clientes, compras, on='id', how='outer')
print('Merge:\n')
unido_outer
````

### Explicación y ejemplo del método concat en pandas

El método `concat` de pandas se utiliza para concatenar (unir) objetos pandas como DataFrames o Series a lo largo de un eje (filas o columnas).

Parámetros clave:
- `objs`: lista o diccionario de objetos pandas a concatenar.
- `axis`: define el eje a lo largo del cual se concatenan los objetos. `axis=0` concatena filas (verticalmente), `axis=1` concatena columnas (horizontalmente).
- `ignore_index`: si es True, se ignoran los índices originales y se crea un nuevo índice secuencial.
- `join`: define cómo se manejan las columnas o índices que no coinciden. Puede ser 'inner' (intersección) o 'outer' (unión, por defecto).

````python
apilado = pd.concat([clientes[['Cliente']], compras[['Compra']]], axis=1)
print('\nConcat (por posición):\n')
apilado
````

````python
apilado = pd.concat([clientes[['Cliente']], compras[['Compra']]], axis=0)
print('\nConcat (por posición):\n')
apilado
````

````python
apilado = pd.concat([clientes[['Cliente']], compras[['Compra']]], axis=0, join="outer", ignore_index=True)
print('\nConcat (por posición):\n')
apilado
````

### 2. Filtrar con query

Permite filtrar con una expresión tipo string; útil para condiciones legibles.

Ejemplo: seleccionar filas de la región Norte con Ventas > 300. 

````python
filtro = ventas.query('Region == "Norte" and Ventas > 500')
print('\nQuery:\n')
filtro
````

> Nota El método query de pandas no acepta SQL completo.

Query permite filtrar un DataFrame usando expresiones similares a las condiciones de Python, con una sintaxis sencilla y legible, pero no es un motor SQL.

Si necesitas ejecutar consultas SQL completas sobre DataFrames, puedes usar librerías como `pandasql` o `duckdb` que permiten ejecutar SQL sobre pandas,

````python
data = {
    'Fecha': pd.date_range(start='2023-01-01', periods=12, freq='ME'),
    'Region': ['Norte', 'Sur', 'Este', 'Oeste'] * 3,
    'Producto': ['A', 'B', 'C'] * 4,
    'Vendedor': ['Juan', 'Ana', 'Luis', 'Marta'] * 3,
    'Ventas': [500, 200, 300, 450, 700, 150, 600, 400, 350, 800, 300, 200],
    'Cantidad': [5, 2, 3, 4, 7, 1, 6, 3, 5, 8, 4, 2],
    'Descuento': [0.05, 0.1, 0.0, 0.07, 0.03, 0.0, 0.1, 0.05, 0.02, 0.0, 0.04, 0.06]
}

df_complejo = pd.DataFrame(data)

df_complejo
````

````python
# 1) Pivot table: suma total de ventas por región y producto
pivot1 = df_complejo.pivot_table(
    index='Region', 
    columns='Producto', 
    values='Ventas', 
    aggfunc='sum', 
    fill_value=0)

print("\nSuma total de ventas por Región y Producto:")
pivot1
````

````python
# 2) Pivot table: promedio de descuento y cantidad por vendedor y región
pivot2 = df_complejo.pivot_table(
    index='Vendedor', 
    columns='Region', 
    values=['Descuento', 'Cantidad'], 
    aggfunc='mean', 
    fill_value=0)

print("\nPromedio de descuento y cantidad por Vendedor y Región:")
pivot2
````

````python
# 3) Pivot table: total de ventas y cantidad por mes (fecha) y producto
pivot3 = df_complejo.pivot_table(
    index='Fecha', 
    columns='Producto', 
    values=['Ventas', 'Cantidad'], 
    aggfunc='sum', 
    fill_value=0)

print("\nTotal de ventas y cantidad por Fecha y Producto:")
pivot3
````
