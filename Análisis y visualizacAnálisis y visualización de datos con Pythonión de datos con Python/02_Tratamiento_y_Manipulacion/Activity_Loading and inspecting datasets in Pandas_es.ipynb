Imagina que eres un desarrollador de Python en una empresa de comercio electrónico en rápido crecimiento. Ellos te han proporcionado un conjunto de datos valioso que contiene información sobre los clientes, sus compras y datos demográficos.

Tu tarea es aprovechar tus habilidades en Python y el poder de la biblioteca Pandas para cargar este conjunto de datos, explorar su estructura y descubrir conocimientos preliminares que guiarán un análisis adicional. En la celda a continuación, comienza importando la biblioteca `pandas` con el alias `pd`. Luego, usa `.read_csv()` para cargar el archivo `customer_data_50.csv` en un DataFrame llamado `customer_data`.

Por último, ejecuta la celda.


```python
# Import the pandas library with the alias 'pd'

# insert code here
import pandas as pd

# Load the CSV file 'customer_data_50.csv' into a DataFrame

# insert code here
customer_data = pd.read_csv('customer_data_50.csv')
```

Ejecute la siguiente celda, que verificará las dimensiones de su DataFrame usando el atributo `.shape`. Esto le indica cuántas filas y columnas tienen sus datos – ¡algo así como averiguar el tamaño de una hoja de cálculo!


```python
# Display the shape of the DataFrame (rows, columns)
print("\nShape of the DataFrame (rows, columns):", customer_data.shape)
```

A continuación, inspeccionarás los datos usando la función `df.head()`, que te permite ver las primeras filas del DataFrame. Esto te da una vista rápida de la estructura y contenido de los datos.

En la celda de abajo, usa `df.head()` para mostrar las primeras 5 filas del DataFrame `customer_data`. Luego, ejecuta la celda y tómate un momento para observar la salida.


```python
# Display the first 5 rows
print("First 5 rows:\n")

# insert code here
print(customer_data.head())
```

Ahora, usarás la función `df.info()`, que proporciona un resumen conciso del DataFrame, incluyendo los nombres de las columnas, sus tipos de datos y el número de valores no nulos.

En la celda de abajo, usa `df.info()` para imprimir información sobre el DataFrame `customer_data`. Luego, ejecuta la celda y tómate un momento para observar la salida.


```python
# Print the column names and their data types
print("\nColumn names and their data types:\n")

# insert code here
print(customer_data.info())
```

A continuación, usarás la función `df.describe()`, que genera estadísticas descriptivas para las columnas numéricas en el DataFrame.

En la celda de abajo, usa `df.describe()` para mostrar estadísticas resumidas para las columnas numéricas en el DataFrame `customer_data`.


```python
# Display descriptive statistics for numerical columns
print("\nDescriptive statistics for numerical columns:\n")

# insert code here
print(customer_data.describe())
```

Finalmente, en la celda de código a continuación, usarás la función `.mean()` en la columna `'age'` de tu `customer_data` para calcular la edad promedio de todos tus clientes.

También usa `.median()` para calcular la edad mediana de todos tus clientes y ejecuta la celda.


```python
# Calculate the mean of the 'age' column
mean_age = customer_data['age'].mean() # insert code here


# Print the mean age
print("\nMean Age:", mean_age)

# Calculate the median of the 'age' column
median_age = customer_data['age'].median() # insert code here

# Print the median age
print("\nMedian Age:", median_age)
```
